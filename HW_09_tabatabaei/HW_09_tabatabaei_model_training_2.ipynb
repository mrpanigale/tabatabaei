# part 2 of diabetic datas question

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
# xgboost
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from collections import Counter
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

```python
# import encoded dataset
extract = pd.read_csv('diabetes+130-us+hospitals+for+years+1999-2008/encoded.csv')
target = extract['readmitted']
extract = extract.drop('readmitted', axis = 1)
diabet_df = pd.concat([extract, target], axis = 1)
display(diabet_df.head())
display(diabet_df.info())

```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(diabet_df.drop('readmitted', axis = 1), diabet_df['readmitted'], test_size = 0.3, random_state = 42)
Xtrain
```

```python
#300 leaves no feature selection
random_forest = RandomForestClassifier(max_leaf_nodes=300,n_estimators=100,random_state=42)
random_forest.fit(Xtrain, ytrain).score(Xtrain, ytrain)
train_score = random_forest.score(Xtrain, ytrain)
test_score = random_forest.score(Xtest, ytest)
train_delta_test = train_score - test_score
print('300 leaves ranodm forest')
print('train score:\n',train_score, '\ntest score:\n',test_score, '\ndelta:\n', train_delta_test)
```

```python
# 100 leaves no feature selection
random_forest = RandomForestClassifier(max_leaf_nodes=100,n_estimators=100,random_state=42)
random_forest.fit(Xtrain, ytrain).score(Xtrain, ytrain)
train_score = random_forest.score(Xtrain, ytrain)
test_score = random_forest.score(Xtest, ytest)
train_delta_test = train_score - test_score
print('100 leaves random forest')
print('train score:\n',train_score, '\ntest score:\n',test_score, '\ndelta:\n', train_delta_test)
```

# faeture selection

```python
# feature selection
main_features = [
    # demographics
    'age',
    'race_Caucasian', 'race_AfricanAmerican', 'race_Hispanic',
    'gender_Male',

    # hospital usage
    'time_in_hospital',
    'number_outpatient', 'number_emergency', 'number_inpatient',

    # treatment intensity
    'num_lab_procedures', 'num_procedures',
    'num_medications', 'number_diagnoses',

    # weight
    'weight_num', 'weight_missing',

    # diagnoses (ICD groups)
    'diag1_Circulatory', 'diag1_Diabetes', 'diag1_Respiratory',
    'diag2_Circulatory', 'diag2_Diabetes',
    'diag3_Circulatory',

    # diabetes tests
    'max_glu_measured', 'max_glu_high',
    'a1c_measured', 'a1c_high',

    # medications
    'insulin_used', 'metformin_used',
    'med_changed', 'diabetesMed',

    # admission / discharge
    'admission_Emergency', 'admission_Elective',
    'discharge_Home', 'discharge_Facility','readmitted'
]

selected_df = diabet_df[main_features]

selected_df.head()
```

```python
X_new_train, X_new_test, y_new_train, y_new_test = train_test_split(selected_df.drop('readmitted', axis = 1), diabet_df['readmitted'], test_size = 0.3, random_state = 42)
```

```python
# 300 leaves with feature selection
random_forest = RandomForestClassifier(max_leaf_nodes=300,n_estimators=100,random_state=42)
random_forest.fit(X_new_train, y_new_train).score(X_new_train, y_new_train)
train_score = random_forest.score(X_new_train, y_new_train)
test_score = random_forest.score(X_new_test, y_new_test)
train_delta_test = train_score - test_score
print('300 leaves ranodm forest')
print('train score:\n',train_score, '\ntest score:\n',test_score, '\ntrain - test:\n', train_delta_test)
```

```python
# 100 leaves with feature selection
random_forest = RandomForestClassifier(max_leaf_nodes=100,n_estimators=100,random_state=42)
random_forest.fit(X_new_train, y_new_train).score(X_new_train, y_new_train)
train_score = random_forest.score(X_new_train, y_new_train)
test_score = random_forest.score(X_new_test, y_new_test)
train_delta_test = train_score - test_score
print('100 leaves ranodm forest')
print('train score:\n',train_score, '\ntest score:\n',test_score, '\ntrain - test:\n', train_delta_test)
```
