```python
import pandas as pd 
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
df =  pd.read_csv("HW_05/iris_dataset.csv")
features = df.columns[:-1]
target = df.columns[-1]
```

step 2

```python
class_0 = df[df[target]==0].copy()
class_1 = df[df[target]==1].copy()
class_2 = df[df[target]==2].copy()

class_0_stats = pd.DataFrame({
    "mean":class_0[features].mean(),
    "var":class_0[features].var(),
    "range":class_0[features].max() - class_0[features].min(),
    "max":class_0[features].max(),
    "min":class_0[features].min()
}).T


class_1_stats = pd.DataFrame({
    "mean":class_1[features].mean(),
    "var":class_1[features].var(),
    "range":class_1[features].max() - class_1[features].min(),
    "max":class_1[features].max(),
    "min":class_1[features].min()
}).T

class_2_stats = pd.DataFrame({
    "mean":class_2[features].mean(),
    "var":class_2[features].var(),
    "range":class_2[features].max() - class_2[features].min(),
    "max":class_2[features].max(),
    "min":class_2[features].min()
}).T
class_0_stats
```

step 3

```python
class_0['petal length prob'] =  norm.pdf(class_0['petal length (cm)'],class_0_stats.loc["mean","petal length (cm)"],np.sqrt(class_0_stats.loc['var','petal length (cm)']))

class_1['petal length prob'] =  norm.pdf(class_1['petal length (cm)'],class_1_stats.loc['mean','petal length (cm)'],np.sqrt(class_1_stats.loc["var",'petal length (cm)']))

class_2['petal length prob'] = norm.pdf(class_2['petal length (cm)'],class_2_stats.loc['mean','petal length (cm)'],np.sqrt(class_2_stats.loc['var','petal length (cm)']))
```

step 4

```python
df['class 0']  = norm.pdf(df['petal length (cm)'],class_0_stats.loc["mean","petal length (cm)"],np.sqrt(class_0_stats.loc['var','petal length (cm)']))
df['class 1'] = norm.pdf(df['petal length (cm)'],class_1_stats.loc['mean','petal length (cm)'],np.sqrt(class_1_stats.loc["var",'petal length (cm)']))
df['class 2'] = norm.pdf(df['petal length (cm)'],class_2_stats.loc['mean','petal length (cm)'],np.sqrt(class_2_stats.loc['var','petal length (cm)']))
df["geuss"] = df[['class 0',"class 1","class 2"]].idxmax(axis=1)
df["geuss"] = df["geuss"].map({"class 0":0,"class 1":1,"class 2":2})
```

step 5

```python
plt.scatter(x=class_0['petal length (cm)'],y=class_0['petal length prob'],label="class 0",color ='blue')
plt.scatter(x=class_1['petal length (cm)'],y=class_1['petal length prob'],label="class 1",color='red')
plt.scatter(x=class_2['petal length (cm)'],y=class_2['petal length prob'],label = "class 2",color="black")
plt.legend()
plt.title("gaussian probs")
plt.xlabel('distribution')
plt.ylabel('gaussi prob')
plt.grid()
plt.show()
```

step 6

```python
# فردا پاشدی با حلقه for و if تعداد پاسخ های درست رو تعیین کن بعد فرمول بزن 
correct_answer,wrong_answer = 0,0
for i in range(len(df)):
    if df.loc[i,target] == df.loc[i,"geuss"] :
        correct_answer += 1
    else:
        wrong_answer+=1

precision = (correct_answer/(correct_answer+wrong_answer))*100

print(f"correct_answer= {correct_answer}")
print(f"wrong_answer= {wrong_answer}")
print(f"precision= {precision}")
df.head()
```
